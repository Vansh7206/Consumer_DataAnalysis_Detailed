## Do Subscribed customer spend more? Compare avg spend and total spend between subs and non subs

```python
subscribers = data[data['Subscription Status'] == 'Yes']
subscribers
```

```python
non_subscribers = data[data['Subscription Status'] == 'No']
non_subscribers
```

```python
avg_spent_subs = subscribers['Purchase Amount (USD)'].mean()
print('Avg spend of Subscribers:','$',avg_spent_subs)
```

```python
avg_spent_nonsubs = non_subscribers['Purchase Amount (USD)'].mean()
print('Avg spend of Non Subscribers:','$',avg_spent_nonsubs)
```

```python
total_subs_spend = subscribers['Purchase Amount (USD)'].sum()
print('Total Spend of Subscribers:','$',total_subs_spend)
```

```python
total_nonsubs_spend = non_subscribers['Purchase Amount (USD)'].sum()
print('Total Spend of Non Subscribers:','$',total_nonsubs_spend)
```

```python
avg_spend = pd.DataFrame({
    'Customer_Type' : ['Subscriber','Non-Subscriber'],
    'Amount $':[avg_spent_subs,avg_spent_nonsubs]
})
avg_total_spend = pd.DataFrame({
    'Customer_Type' : ['Subscriber','Non-Subscriber'],
    'Amount $':[total_subs_spend,total_nonsubs_spend]
})
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 5))

# LEFT PLOT â€” Average Spend
avg_spend.plot(kind='bar', x='Customer_Type', y='Amount $', ax=axs[0])
axs[0].set_title('Avg Spend Subs vs Non-Subs')

# RIGHT PLOT â€” Total Spend
avg_total_spend.plot(kind='bar', x='Customer_Type', y='Amount $', ax=axs[1])
axs[1].set_title('Total Spend Subs vs Non-Subs')

plt.tight_layout()
plt.show()
```

## Segement customer based upon new,returning,loyal based on their total number of previous purchases and show count of each segment

```python
def my_customer(purchase):
    if purchase <= 1:
        return 'New'
    elif purchase <= 10:
        return 'Returning'
    else:
        return 'Loyal'

data['count_customer'] = data['Previous Purchases'].apply(my_customer)
```

```python
final = data['count_customer'].value_counts()
final
```

```python
final.plot(kind='bar', figsize=(5,4))
plt.title("Customer Segmentation by Purchase History")
plt.ylabel("Count")
plt.xlabel("Segment")
plt.show()
```
